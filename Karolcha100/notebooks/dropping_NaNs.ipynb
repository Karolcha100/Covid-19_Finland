```python
import pandas as pd
```

```python
df_raw = pd.read_csv("../../data_raw/SG.csv")

df_raw
```

```python
df_raw["date"] = pd.to_datetime(df_raw["date"])

df_raw
```

```python
df_no_str = df_raw.select_dtypes(exclude='string')

df_no_str = df_no_str.set_index("date")

df_no_str
```

```python
for col in df_no_str.columns:
    print(col) if "search" not in col else None
```

```python
df_sel_cols = df_no_str.drop(columns = ["openstreetmap_id", "latitude", "longitude", "openstreetmap_id", "aggregation_level"])

df_sel_cols = df_sel_cols[:-5]

df_sel_cols
```

```python
df_final = df_sel_cols.interpolate(method="linear").fillna(0)

df_final
```

```python
df_final["search_trends_abdominal_pain"]
```
